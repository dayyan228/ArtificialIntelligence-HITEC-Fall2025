```python
def factorial(x):
    if(x<2):
        return 1
    else:  
        return (x * factorial(x-1))

value = factorial(7)

print(value)
```

```python
def findFactorial(num):
    if num == 1: 
        return 1
    else: 
        return num*(findFactorial(num-1))
```

```python
number = 12

result = findFactorial(number)
print(f"The factorial of: {number} is {result}")
```

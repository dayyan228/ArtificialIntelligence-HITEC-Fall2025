```python
class Environment:
    def __init__(self):
        self.rooms = {'A': 'Dirty', 'B': 'Dirty', 'c': 'Dirty'}
        self.agent_location = 'A'	

    def get_percept(self):
        return self.agent_location, self.rooms[self.agent_location]

    def execute_action(self, action):
        if action == 'Vacuum':
            self.rooms[self.agent_location] = 'Clean'
        elif action == 'Right':
            self.agent_location = 'B'
        elif action == 'Left':
            self.agent_location = 'A'

class SimpleReflexAgent:
    def __init__(self):
        self.rules = {
            ('A', 'Dirty'): 'Vacuum',
            ('B', 'Dirty'): 'Vacuum',
            ('A', 'Clean'): 'Right',
            ('B', 'Clean'): 'Left'
        }

    def select_action(self, percept):
        return self.rules.get(percept, 'NoOp')

# Simulation
env = Environment()
agent = SimpleReflexAgent()

for step in range(5):
    percept = env.get_percept()
    action = agent.select_action(percept)
    print(f"Step {step + 1}: Percept: {percept}, Action: {action}")
    env.execute_action(action)
```

```python
import cv2
import random

# Initialize webcam (sensor)
camera = cv2.VideoCapture(0)

# Load Haar Cascade face detection model (internal knowledge)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

print("ðŸŽ¥ Simple Reflex Agent Started (Press 'q' to Quit)")

while True:
    ret, frame = camera.read()
    if not ret:
        break

    # Convert to grayscale for detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40))

    # Reflex logic: For each face â†’ draw a colored rectangle
    for (x, y, w, h) in faces:
        color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))  # unique color
        cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
        cv2.putText(frame, "Face Detected", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # Display the frame (actuator output)
    cv2.imshow("Simple Reflex Agent - Multi Face Detection", frame)

    # Break loop on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```

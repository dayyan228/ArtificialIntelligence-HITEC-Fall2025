```python
import numpy as np
from collections import deque

# Step 1: Generate 50 random integers between 5 and 500
random_numbers = np.random.randint(5, 501, size=50)
print("Generated Numbers:\n", random_numbers)
print("\n")

# Step 2: Define a Binary Tree Node class
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Step 3: Insert values into Binary Search Tree
def insert(root, value):
    if root is None:
        return Node(value)
    if value < root.value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    return root

# Step 4: Build the BST using the random numbers
root = None
for num in random_numbers:
    root = insert(root, num)

# Step 5: Perform Breadth First Search (Level Order Traversal)
def bfs_traversal(root):
    if not root:
        return []
    queue = deque([root])
    result = []
    while queue:
        current = queue.popleft()
        result.append(current.value)
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
    return result

# Step 6: Display the BFS Traversal Output
bfs_result = bfs_traversal(root)
print("BFS Traversal of Generated Tree:")
print(bfs_result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from collections import deque

# Step 1: Generate 50 random integers between 5 and 500
random_numbers = np.random.randint(5, 501, size=50)
print("Generated Numbers:\n", random_numbers)
print("\n")

# Step 2: Define a Binary Tree Node
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Step 3: Insert values into Binary Search Tree
def insert(root, value):
    if root is None:
        return Node(value)
    if value < root.value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    return root

# Step 4: Build the BST
root = None
for num in random_numbers:
    root = insert(root, num)

# Step 5: Perform Breadth First Search (Level Order Traversal)
def bfs_traversal(root):
    if not root:
        return []
    queue = deque([root])
    result = []
    while queue:
        current = queue.popleft()
        result.append(current.value)
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
    return result

bfs_result = bfs_traversal(root)
print("BFS Traversal of Tree:")
print(bfs_result)

# Step 6: Build graph for visualization
def build_graph(G, node, pos, x=0, y=0, layer=1):
    if node is not None:
        G.add_node(node.value, pos=(x, y))
        if node.left:
            G.add_edge(node.value, node.left.value)
            build_graph(G, node.left, pos, x - 1 / layer, y - 1, layer + 1)
        if node.right:
            G.add_edge(node.value, node.right.value)
            build_graph(G, node.right, pos, x + 1 / layer, y - 1, layer + 1)

# Step 7: Create and plot the tree
G = nx.DiGraph()
pos = {}
build_graph(G, root, pos)

positions = nx.get_node_attributes(G, 'pos')
plt.figure(figsize=(14, 8))
nx.draw(
    G,
    pos=positions,
    with_labels=True,
    arrows=False,
    node_size=800,
    node_color="lightblue",
    font_size=10,
    font_weight="bold"
)
plt.title("Binary Search Tree Visualization", fontsize=14)
plt.show()
```

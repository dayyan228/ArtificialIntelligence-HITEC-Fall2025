```python
import random
import time

def generate_unique_random_float(start: float, end: float) -> float:
    seed = time.time_ns()
    random.seed(seed)
    return random.uniform(start, end)

def generate_random_float_list(start: float, end: float, count: int) -> list:
    random_floats = []
    for _ in range(count):
        rand_num = generate_unique_random_float(start, end)
        random_floats.append(rand_num)
        time.sleep(0.001)  # Sleep for 1ms to ensure unique seed
    return random_floats

# Example usage
start_range = 10.5
end_range = 19.9
number_count = 10

float_list = generate_random_float_list(start_range, end_range, number_count)
print("Generated list of unique floats:")
print(float_list)
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = round(value, 2)
        self.left = None
        self.right = None

def build_perfect_tree(values):
    if not values:
        return None
    mid = len(values) // 2
    root = TreeNode(values[mid])
    root.left = build_perfect_tree(values[:mid])
    root.right = build_perfect_tree(values[mid+1:])
    return root
```

```python
import matplotlib.pyplot as plt

def plot_tree(node, x=0, y=0, dx=1.0, dy=1.5, ax=None, depth=0):
    if node is None:
        return
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.axis('off')

    ax.text(x, y, f"{node.value:.2f}", ha='center', va='center',
            bbox=dict(boxstyle="circle", facecolor="lightblue"))

    if node.left:
        ax.plot([x, x - dx], [y, y - dy], 'k-')
        plot_tree(node.left, x - dx, y - dy, dx / 1.5, dy, ax, depth + 1)
    if node.right:
        ax.plot([x, x + dx], [y, y - dy], 'k-')
        plot_tree(node.right, x + dx, y - dy, dx / 1.5, dy, ax, depth + 1)

    if depth == 0:
        plt.tight_layout()
        plt.savefig("matplotlib_tree.png")
        plt.show()
```

```python
import random
import time

def generate_random_float_list(start, end, count):
    floats = []
    for _ in range(count):
        seed = time.time_ns()
        random.seed(seed)
        floats.append(round(random.uniform(start, end), 2))
        time.sleep(0.001)
    return floats

def adjust_to_perfect_size(values):
    import math
    h = int(math.log2(len(values) + 1))
    return values[:2**h - 1]

# Generate and build
float_list = generate_random_float_list(10.5, 99.9, 15)
perfect_list = adjust_to_perfect_size(sorted(float_list))
tree_root = build_perfect_tree(perfect_list)

# Plot and save
plot_tree(tree_root)
```

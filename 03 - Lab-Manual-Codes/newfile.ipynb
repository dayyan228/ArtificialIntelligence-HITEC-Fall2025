```python
import os, sys
current_limit = sys.getrecursionlimit()
print(f"The current recursion limit is: {current_limit}")


sys.setrecursionlimit(99999999)
current_limit = sys.getrecursionlimit()
print(f"The current recursion limit is: {current_limit}")

def func(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return (n+func(n-1)+func(n-2))
    
print(func(12))
```

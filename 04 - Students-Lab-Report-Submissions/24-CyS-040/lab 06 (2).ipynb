```python
from collections import deque
import heapq

graph1 = { 
    0: [1, 2], 
    1: [3, 4], 
    2: [5, 6], 
    3: [], 
    4: [], 
    5: [], 
    6: [] 
}

print("\n================ GRAPH PROBLEM 1 ================")

# ---------------------- BFS ----------------------
def bfs(graph, start):
    print("\nBFS Traversal:")
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            queue.extend(graph[node])

# ---------------------- DFS ----------------------
def dfs(graph, start):
    print("\nDFS Traversal:")
    visited = set()
    stack = [start]
    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            stack.extend(reversed(graph[node]))

# ---------------------- DLS ----------------------
def dls(graph, node, limit, depth=0):
    print(node, end=" ")
    if depth == limit:
        return
    for child in graph[node]:
        dls(graph, child, limit, depth + 1)

# ---------------------- IDDFS ----------------------
def iddfs(graph, start, max_depth):
    print("\nIDDFS Traversal:")
    for depth in range(max_depth + 1):
        print(f"\nDepth {depth}: ", end="")
        dls(graph, start, depth)

# ---------------------- UCS ----------------------
def ucs(graph, start):
    print("\nUCS Traversal:")
    visited = set()
    queue = [(0, start)]
    while queue:
        cost, node = heapq.heappop(queue)
        if node in visited:
            continue
        print(f"{node}(cost={cost})", end=" ")
        visited.add(node)
        for child in graph[node]:
            heapq.heappush(queue, (cost + 1, child))

# ---------------------- RUN ALL ----------------------
bfs(graph1, 0)
dfs(graph1, 0)
print("\n\nDLS (Depth Limit = 2):")
dls(graph1, 0, 2)
iddfs(graph1, 0, 3)
ucs(graph1, 0)
```

```python
num = int(input("Enter the number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
list1 = ['abc', 'xyz', 'aba', '1221', 'xyzzyx', 'aa', '122']
count = 0

for word in list1:
    if len(word) >= 2 and word[0] == word[-1]:
        count = count + 1

print("Expected Result:", count)
```

```python
numbers = (1, 2, 4, 2, 6)
sum_numbers = 0
product = 1

for n in numbers:
    sum_numbers = sum_numbers + n
    product = product * n

print("Sum is:", sum_numbers)
print("Product is:", product)
```

```python
n = int(input("Enter a number: "))
d = {}

for x in range(1, n + 1):
    d[str(x)] = x * x

print(d)
```

```python
import math 
class Circle: 
    def __init__(self, radius): 
        self.radius = radius 
    def area(self): 
        return math.pi * self.radius ** 2 
    def perimeter(self): 
        return 2 * math.pi * self.radius 
c = Circle(5) 
print(c.area()) 
print(c.perimeter())
```

```python
class Power: 
    def pow(self, x, n): 
        result = 1 
        if n == 0: 
            return 1 
        elif n > 0: 
            for i in range(n): 
                result *= x 
        else: 
            for i in range(-n): 
                result *= x 
            result = 1 / result 
        return result 
 
p = Power() 
print(p.pow(2, 3)) 
print(p.pow(2, -3))
```

```python
class Item:
    def __init__(self, name, description, price):
        self.name = name
        self.description = description
        self.price = price

    def viewFullDescription(self):
        print("Name:", self.name)
        print("Description:", self.description)
        print("Price:", self.price)

    def addToShoppingBasket(self):
        print(self.name, "added to shopping basket")

    def removeFromShoppingBasket(self):
        print(self.name, "removed from shopping basket")


class MP3(Item):
    def __init__(self, name, description, price, artist, duration):
        super().__init__(name, description, price)
        self.artist = artist
        self.duration = duration

    def play(self):
        print("Playing", self.name, "by", self.artist)

    def download(self):
        print("Downloading", self.name)


class DVD(Item):
    def __init__(self, name, description, price, certificate, duration, actors):
        super().__init__(name, description, price)
        self.certificate = certificate
        self.duration = duration
        self.actors = actors

    def viewTrailer(self):
        print("Viewing trailer for", self.name)


class Book(Item):
    def __init__(self, name, description, price, author, numberOfPages, genre):
        super().__init__(name, description, price)
        self.author = author
        self.numberOfPages = numberOfPages
        self.genre = genre

    def previewContent(self):
        print("Previewing content of", self.name, "by", self.author)


# Create objects and test
m = MP3("Song A", "Hit single", 500000, "Artist X", "3:45")
m.viewFullDescription()
m.play()
m.download()

d = DVD("Movie B", "Action movie", 1000000, "PG-13", "2h", ["Actor 1", "Actor 2"])
d.viewFullDescription()
d.viewTrailer()

b = Book("Book C", "Fantasy novel", 7000, "Author Y", 320, "Fantasy")
b.viewFullDescription()
b.previewContent()
```

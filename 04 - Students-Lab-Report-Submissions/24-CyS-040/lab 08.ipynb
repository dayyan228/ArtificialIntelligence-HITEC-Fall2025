```python
import math 
def alpha_beta(depth, nodeIndex, isMax, scores, alpha, beta, targetDepth): 
    # Terminal node 
    if depth == targetDepth: 
        return scores[nodeIndex] 
    if isMax: 
        best = -math.inf 
        for i in range(2): 
            value = alpha_beta(depth + 1, nodeIndex * 2 + i, 
                               False, scores, alpha, beta, targetDepth) 
            best = max(best, value) 
            alpha = max(alpha, best) 
            if beta <= alpha: 
                break   # Beta cut-off 
        return best 
    else: 
        best = math.inf 
        for i in range(2): 
            value = alpha_beta(depth + 1, nodeIndex * 2 + i, 
                               True, scores, alpha, beta, targetDepth) 
            best = min(best, value) 
            beta = min(beta, best) 
            if beta <= alpha: 
                break   # Alpha cut-off 
        return best 
# Leaf node values 
scores = [4, 2, -3, -6, 7, 0, 5, 8] 
treeDepth = int(math.log2(len(scores))) 
result = alpha_beta(0, 0, True, scores, -math.inf, math.inf, treeDepth) 
print("The optimal value using Alpha-Beta Pruning is:", result)
```

```python
import math

PLAYER = 'x'
OPPONENT = 'o'

def is_moves_left(board):
    for row in board:
        if '_' in row:
            return True
    return False

def evaluate(board):
    # Check rows
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != '_':
            return 10 if board[i][0] == PLAYER else -10
    # Check columns
    for i in range(3):
        if board[0][i] == board[1][i] == board[2][i] != '_':
            return 10 if board[0][i] == PLAYER else -10
    # Check diagonals
    if board[0][0] == board[1][1] == board[2][2] != '_':
        return 10 if board[0][0] == PLAYER else -10
    if board[0][2] == board[1][1] == board[2][0] != '_':
        return 10 if board[0][2] == PLAYER else -10
    return 0

def minimax(board, is_max):
    score = evaluate(board)
    
    # If someone has won
    if score == 10 or score == -10:
        return score
    
    # If no moves left, it's a tie
    if not is_moves_left(board):
        return 0
    
    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = PLAYER
                    best = max(best, minimax(board, False))
                    board[i][j] = '_'
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = OPPONENT
                    best = min(best, minimax(board, True))
                    board[i][j] = '_'
        return best

# Example board
board = [
    ['o', 'x', 'o'],
    ['o', 'x', 'x'],
    ['x', 'o', 'x']
]

print("Minimax evaluation:", minimax(board, True))
```

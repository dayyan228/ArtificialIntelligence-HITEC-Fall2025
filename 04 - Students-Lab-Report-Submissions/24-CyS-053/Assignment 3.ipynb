```python
def string_match(string1,string2):
    print(f"String Length is: {len(string1)}")
    print(f"String Length is: {len(string2)}")
    if len(string1)==len(string2):
        if string1==string2:
            print("Both Strings are Same")
        else:
            print("Same Length But not the Same")
    else:
        print("Not the Same length")
string1 = input("Enter First String")
string2 = input("Enter Second String")
string_match(string1,string2)
```

```python
def factorial(num):
    if num<0:
        return "Error: Wrong Number"
    elif num ==0 or num ==1:
        return 1
    else:
        result =1
        for i in range(1,num+1):
            result *=i
    return result

print(factorial(4))
```
